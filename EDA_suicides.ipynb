```python
# descripcion de variables
#ParentLocation= continente
#Location al país
#Period al año, Dim1 al sexo 
#FactValueNumeric corresponde a la tasa de suicidio normalizada por 100.000 habitantes. 


new_df= df[["ParentLocation", "Location", "Period","Dim1","FactValueNumeric","Value"]]
```

```python
# realizamos en analisis  solamente a la categoría ambos sexos (“Both sexes”) 

df_analisis=new_df[new_df["Dim1"]=="Both sexes"]
df_analisis
```

```python
#ordenamos df por FactValueNumeric
df_analisis.sort_values("FactValueNumeric", ascending=False).head(10)

#top ten de las mayores tasas de suicidio en (2000-2019) es Lesoto - Africa.
```

```python
# paises con mayor tasa de suicidio en el año 2019
df_analisis2019= df_analisis[df_analisis["Period"]==2019]
top10MAX=df_analisis2019.sort_values("FactValueNumeric", ascending= False).head(10)
top10MAX

# 6 de los 10 paises pertecenen a AFRICA
```

```python
# paises con MENOR tasa de suicidio en el año 2019
df_analisis2019= df_analisis[df_analisis["Period"]==2019]
top10MIN=df_analisis2019.sort_values("FactValueNumeric").head(10)
top10MIN

# 6 de los 10 paises ubicados en AMERICA
```

```python
# Analisis por continente (2000-2019)

df_cont=new_df.groupby("ParentLocation")["FactValueNumeric"].agg([min,max,np.mean]).sort_values("mean", ascending=False)
df_cont
```

```python
#graficamos

df_cont.plot(kind="bar",
             xlabel="Tasa de suicidio / 100.000 habitantes",
             ylabel="Continente",
            title="Análisis de tasa de suicidio por continente (OMS) 2000-2019")
```

```python
df_cont_mean=new_df.groupby("ParentLocation")["FactValueNumeric"].agg(np.mean)
df_cont_mean
```

```python
#graficamos

df_cont_mean.plot(kind="bar", 
                  xlabel="Tasa de suicidio / 100.000 habitantes",
                  ylabel="Continente",
                  title="Tasa de suicidio promedio por continente (OMS) 2000-2019")

```

```python
#Tabla pivote

df_pivot=df_analisis.pivot_table(index="ParentLocation", columns="Period", values="FactValueNumeric")
df_pivot
```

```python
#Visualización de curvas de tasas por continente y año.

#1.  df por continente
df_AM=df_analisis[df_analisis["ParentLocation"]=="Americas"]
df_AF=df_analisis[df_analisis["ParentLocation"]=="Africa"]
df_EMED=df_analisis[df_analisis["ParentLocation"]=="Eastern Mediterranean"]
df_EU=df_analisis[df_analisis["ParentLocation"]=="Europe"]
df_SEA=df_analisis[df_analisis["ParentLocation"]=="South-East Asia"]
df_WPAC=df_analisis[df_analisis["ParentLocation"]=="Western Pacific"]

df_AM
```

```python
#2. Agrupamos por continente el periodo y FactValueNumeric

df_AM_group=df_AM.groupby("Period")["FactValueNumeric"].mean().reset_index()
df_AF_group=df_AF.groupby("Period")["FactValueNumeric"].mean().reset_index()
df_EMED_group=df_EMED.groupby("Period")["FactValueNumeric"].mean().reset_index()
df_EU_group=df_EU.groupby("Period")["FactValueNumeric"].mean().reset_index()
df_SEA_group=df_SEA.groupby("Period")["FactValueNumeric"].mean().reset_index()
df_WPAC_group=df_WPAC.groupby("Period")["FactValueNumeric"].mean().reset_index()

df_AM_group
```

```python
# Graficamos curvas

plt.figure()
plt.plot(df_AM_group["Period"],df_AM_group["FactValueNumeric"], label="América")
plt.plot(df_AF_group["Period"],df_AF_group["FactValueNumeric"], label="Africa")
plt.plot(df_EMED_group["Period"],df_EMED_group["FactValueNumeric"], label="Eastern Mediterranean")
plt.plot(df_EU_group["Period"],df_EU_group["FactValueNumeric"], label="Europe")
plt.plot(df_SEA_group["Period"],df_SEA_group["FactValueNumeric"], label="South-East Asia")
plt.plot(df_WPAC_group["Period"],df_WPAC_group["FactValueNumeric"], label="Western Pacific")
plt.title("Tasa de Suicidio por 100.000 habitantes según regiones (OMS).2000-2019",
         loc="center",
         size=10)
plt.xlabel("Años",
          labelpad=10)
plt.ylabel("Tasa Cruda de Suicidio por 100.000 hbtes",
          labelpad=10)
```
